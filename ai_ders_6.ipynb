# Bulanık Mantık : (Zadeh ,1965)
- Yapay zekanın bir alt konusu olan derin öğrenme ile imge işleme
konusunda çok başarılı sonuçlar alırken, diğer bir alt kolu olan
makine öğrenmesi ile veri bilimi çalışmaları gerçekleştirilebilmektedir.Bulanık mantık ile kontrol sistemleri üzerinde geliştirilen uygulamalarda karar verme çalışmaları gerçekleştirilmektedir.

- Bulanık küme öğesi aynı anda birden fazla aynı değişken özelliğine sahip başka bir kümenin de elemanı olabilmektedir. Bulanık üyelik kavramı sözel ifadelerde oluşturulup belirli bir sayısal aralığı ifade etmektedir

- #### **Bulanık mantık, içerisinde kullanılan fonksiyonların farklı sözel kavramlarla ifade edilmesi prensibine göre çalışmaktadır.**

### Küme Kuramı :
- Bulanık kümeler ile tasarlanacak olan sistemin parametreleri ortaya çıkarılmaktadır. Bu parametrelerin belirlenmesindeki amaç ilgili sistemin ileri aşamalarında sonucun daha başarılı olmasında rol oynamaktadır.

- **Bulanık kurallar “eğer- o halde” kuralları ile oluşturulmaktadır. Bu kuralları oluştururken kuralları birbirine bağlar iken “Ve”, “Veya” ve “Değil” tanımları kullanılmaktadır.**


- A kümesine dahil elemanların üyelik derecesi 1,
- A kümesine dahil olmayan elemanların üyelik derecesi 0 olarak alınır.

# ______________ ___________ _______________ ________ _______ __________________ ______ ___________ ___________ _______________ ____________ ___________ ____________ _______________ ____ ____________ _________ ___________ _______ ________ _____ ____________

### Üyelik Derecelerinin Belirlenmesi :
Her bir elemanın kendine ait bir üyelik derecesi ile ifade edildiğinden bahsettik. Peki bu üyelik derecelerini nasıl belirlenir. Herhangi bir değer aralığındaki elemanlar için bir kümeye hangi dereceden ait olduklarını gösteren fonksiyonlara üyelik fonksiyonları (membership function) denir. µF şeklinde belirtilir. Bu fonksiyonları ifade etmek içinse genelde üçgen, yamuk, çan eğrisi, üstel, gaussian gibi fonksiyonlar kullanılır.

```python
import pandas as pd
import skfuzzy as fuzz
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print("Python version : ", sys.version)
print("Pandas version : ", pd.__version__)
print("Matplotlib version : ", matplotlib.__version__)
print("Skfuzzy version : ", fuzz.__version__)
print("Numpy version : ", np.__version__)
```

## *garson - bahşiş uygulaması*



## arange() Fonksiyonu Nedir?
- Numpy dizisi oluşturur. ( array )
- Başlangıç / Start , Bitiş / Stop , Adım / Step (opsiyonel)


```python
## -- 0'dan 22'e kadar 1'er arttırmak (0,23,1)
service = np.arange(0, 23, 1) ## -- giriş değişkeni
meal = np.arange(0, 23, 1) ## -- giriş değişkeni
tip = np.arange(0, 231, 1) ## çıkış değişkeni
```

## Bulanık Sistem Tasarımı :
- trimf = üçgen üyelik fonksiyonu
- trumf = yamuk üyelik fonksiyonu

```python
BQS = fuzz.trimf(service, [0, 0, 11]) ## -- BQS = servis kalitesi kötü
## -- ^ 0'dan 11'e kadar kadar olan tüm değerlendirmeler kötü kalite demektir.

MQS = fuzz.trimf(service, [0, 11, 22]) ## -- MQS = servis kalitesi orta
## -- ^ 0'dan 11'e kadar kötü 11'den de 22'ye kadar orta kalite olarak kalıcaktır ve 22'ye geldıgınde sürecini bitirirr

GQS = fuzz.trimf(service, [11, 22, 22]) ## -- GQS = servis kalitesi iyi
## -- ^ 11'den 22'ye kadar iyi kalite de oldugunu gösterir.
```

```python
BM = fuzz.trimf(meal, [0, 0, 11]) ## -- BM = yemek kötü
MM = fuzz.trimf(meal, [0, 11, 22]) ## -- MM = yemek orta
GM = fuzz.trimf(meal, [11, 22, 22]) ## -- GM = yemek iyi
```

```python
VBT = fuzz.trimf(tip, [0, 0, 30]) ## -- bahşiş cok kötü
BT = fuzz.trimf(tip, [20, 50, 80]) ## -- bahşiş kötü
MT = fuzz.trimf(tip, [70, 100, 130]) ## -- bahşiş orta
GT = fuzz.trimf(tip, [120, 150, 180]) ## -- bahşiş iyi
VGT = fuzz.trimf(tip, [170, 230, 230]) ## -- bahsis cok iyi
```

### Hizmet kalite grafigi :

```python
plt.figure(figsize=(18, 9)) ## -- Grafiğin genişliği 18 birim, yüksekliği ise 9 birim.
## -- linewidth=2.4 =  Çizgilerin kalınlığı
plt.plot(service, BQS, 'r', linewidth=2.4, label='BQS = Bad Quality Service')
plt.plot(service, MQS, 'b', linewidth=2.4, label='MQS = Medium Quality Service')
plt.plot(service, GQS, 'g', linewidth=2.4, label='GQS = Good Quality Service')
plt.title('Service Quality') ## -- grafik basligi
plt.legend() ## -- kutucuk
plt.tight_layout() ## -- grafik elemanlarının iç içe geçmesini önler
plt.show()
```

### Yemek Kalite Grafigi :

```python
plt.figure(figsize=(18, 9))
plt.plot(meal, BM, 'r', linewidth=2.4, label='BM = Bad Meal')
plt.plot(meal, MM, 'b', linewidth=2.4, label='MM = Medium Meal')
plt.plot(meal, GM, 'g', linewidth=2.4, label='GM = Good Meal')
plt.title('Meal Quality')
plt.legend()
plt.tight_layout()
plt.show()
```

### Bahşiş Miktarı Grafigi :

```python
plt.figure(figsize=(18, 10))
plt.plot(tip, VBT, 'brown', linewidth=4, label='VBT = Very Bad Tip')
plt.plot(tip, BT, 'orangered', linewidth=4, label='BT = Bad Tip')
plt.plot(tip, MT, 'b', linewidth=4, label='MT = Medium Tip')
plt.plot(tip, GT, 'greenyellow', linewidth=4, label='GT = Good Tip')
plt.plot(tip, VGT, 'forestgreen', linewidth=4, label='VGT = Very Good Tip')
plt.title('Tip Amount')
plt.legend()
plt.tight_layout()
plt.show()
```

### Kullanıcı Girişine Göre Bahşiş Değerlendirme

```python
## -- kullanıcı girişi
service_point = input('Rate service quality from 0 to 22 : ')
meal_point = input('Rate food quality from 0 to 22 : ')
```

**fuzz.interp_membership FONKSIYONU :**

- Verilen bir değerin belirtilen fuzzy kümede, ne kadar yer aldığını hesaplar ve bu değerin *üyelik derecesini* döndürür.

```python
s_bad = fuzz.interp_membership(service, BQS, float(service_point))
s_mid = fuzz.interp_membership(service, MQS, float(service_point))
s_good = fuzz.interp_membership(service, GQS, float(service_point))

m_bad = fuzz.interp_membership(meal, BM, float(meal_point))
m_mid = fuzz.interp_membership(meal, MM, float(meal_point))
m_good = fuzz.interp_membership(meal, GM, float(meal_point))
```

`Kurallar oluşturuluyor :`

**np.fmin FONKSIYONU :**

- Kuralların üyelik derecelerinin *kesişimini* alır.

```python
rule_1 = np.fmin(s_bad, m_bad) ## -- (Kötü Servis ve Kötü Yemek):
control_rule_1 = np.fmin(rule_1, VBT)

rule_2 = np.fmin(s_bad, m_mid) ## -- (Kötü Servis ve Orta Yemek):
control_rule_2 = np.fmin(rule_2, BT)

rule_3 = np.fmin(s_mid, m_mid) ## -- (Orta Servis ve Orta Yemek)
control_rule_3 = np.fmin(rule_3, MT)

rule_4 = np.fmin(s_mid, m_good) ## -- (Orta Servis ve İyi Yemek)
control_rule_4 = np.fmin(rule_4, GT)

rule_5 = np.fmin(s_good, m_bad) ## -- (İyi Servis ve Kötü Yemek)
control_rule_5 = np.fmin(rule_5, MT)

rule_6 = np.fmin(s_good, m_good) ## -- (İyi Servis ve İyi Yemek)
control_rule_6 = np.fmin(rule_6, VGT)
```

```python
ctr0 = np.zeros_like(tip) ## -- np_zeros_like = Bahşiş Kontrol hesaplamalarına başlamadan önce 0'ı başlangıç noktası belirler.

## -- Kodun temel amacı, her bir kuraldan elde edilen sonuçları yani kontrol değerlerini en güçlü sonuç bulmak için maksimum (fmax) olarak alır
c1 = np.fmax(control_rule_1, control_rule_2)
c2 = np.fmax(control_rule_3, control_rule_4)
c3 = np.fmax(control_rule_5, control_rule_6)

c4 = np.fmax(c2, c3)

toplanan_kurallar = np.fmax(c1,c4)
```

 **fuzz.defuzz FONKSIYONU :**
- Bulanık kümelerin (fuzzy sets) defuzzification işlemi için kullanılır.  
- Bulanık sonuçları bir sayı (kesin sonuç) haline getirir.



### ____________ __________ _________ ________ ___________ ____________ _________  ____________ ____ __________ ______________ ___________ ________ ____________ ________ ________ ________ _______

## defuzzification yöntemleri :

###**1-) Centroid Yöntemi :**
- Bulanık kümeyi bir ağırlıklı ortalama kullanarak keskin bir değere dönüştürür. Kümelerin "merkezini" bulur.

###**2-) Weighted Average (Ağırlıklı Ortalama) :**
- Her değerin üyelik derecesiyle çarpıldığı ve sonra tüm bu çarpımların ortalamasının alındığı bir yöntemdir. Bu, Centroid'e benzer bir yaklaşımdır.

###**3-) Largest of Maximum (LOM) - Maksimumların En Büyüğü :**
- Bulanık kümedeki en yüksek üyelik derecesine sahip olan elemanı bulur ve bu değeri netleştirilmiş sonuç olarak alır.

###**4-) Mean of Maximum (MOM) - Maksimumların Ortalaması :**
- Bulanık kümenin en yüksek üyelik derecesine sahip olan elemanların ortalamasını alır.

###**5-) Smallest of Maximum (SOM) - Maksimumların En Küçüğü :**
- Bulanık kümenin en yüksek üyelik derecesine sahip olan elemanların en küçük olanını alır.

###**6-) Bisector Yöntemi :**
- Kümeyi iki eşit bölüme ayıran noktayı bulur.




```python
## -- fuzz.defuzz() fonksiyonu, bulanık sonuçları, kesin değere dönüştürmek için kullanılır.
durulama_islemi = fuzz.defuzz(tip, toplanan_kurallar, 'centroid')
islem_sonucu = fuzz.interp_membership(tip, toplanan_kurallar, durulama_islemi)
```

```python
fig, ax4 = plt.subplots(figsize=(17,9))

ax4.plot(tip, VBT, 'brown', linewidth=1.5, linestyle='--')
ax4.plot(tip, BT, 'r',  linewidth=1.5, linestyle='--')
ax4.plot(tip, MT, 'peru',  linewidth=1.5, linestyle='--')
ax4.plot(tip, GT, 'b',  linewidth=1.5, linestyle='--')
ax4.plot(tip, VGT, 'forestgreen', linewidth=1.5, linestyle='--')

ax4.fill_between(tip, ctr0, toplanan_kurallar, facecolor='lightskyblue', alpha=0.7)
ax4.plot([durulama_islemi, durulama_islemi], [0, islem_sonucu], 'k', linewidth=1.5, alpha=0.9)
ax4.set_title('Result')
plt.tight_layout()
plt.show()

print("Tip Amount = ", durulama_islemi, " TL")
```
