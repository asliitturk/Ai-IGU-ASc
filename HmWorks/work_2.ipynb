```python
import cv2
import matplotlib.pyplot as plt
```

```python
folder_path = '/content/drive/My Drive/Ai_Ders_IGU_2024/pictures'
pic_formats = ('.jpg', '.jpeg', '.png', '.gif', '.bmp')
```

```python
## -- orijinal görüntü yükleme ve gösterme
image = cv2.imread('pictures/trafik-1.png')

plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Display of uploaded image")
plt.show()
```

```python
## -- gri tonlama
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow(gray_image , cmap = 'gray')
plt.title("Display of uploaded image")
plt.show()
```

```python
## -- bulanıklaştırma
blurred_image = cv2.GaussianBlur(gray_image, (5, 5), 0)

plt.imshow(blurred_image, cmap = 'gray')
plt.title("Blurred image")
plt.show()
```

```python
## -- kenar algılama
edges = cv2.Canny(blurred_image, threshold1=100, threshold2=200)

plt.imshow(edges, cmap='gray')
plt.title("Edge Detection")
plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.subplot(1, 4, 1)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")

plt.subplot(1, 4, 2)
plt.imshow(gray_image, cmap='gray')
plt.title("Grayscale")

plt.subplot(1, 4, 3)
plt.imshow(blurred_image, cmap='gray')
plt.title("Blurred Image")

plt.subplot(1, 4, 4)
plt.imshow(edges, cmap='gray')
plt.title("Edge Detection")

plt.show()
```
