```python
import pandas as pd ## -- pandas kütüphanesi & alias'ı (pd)
## -- Caroline Herschel'in ders ve not bilgileri
data = {'Lesson' : ['Stellar Evolution','High Energy Astrophysics','Long Wave Astronomy','Practical Astronomy'],
        'Score' : [100,85,99,89] }
```

```python
df = pd.DataFrame(data) ## -- veriyi bir DataFrame'e dönüştürür
## -- df: Bu, oluşturulan DataFrame'in depolandığı değişkendir.
## -- Artık bu değişkeni kullanarak veri çerçevesine erişebilir ve onun üzerinde işlemler gerçekleştirebiliriz.
```

```python
highest_score = df['Score'].max() ## -- max score = en yük not
print("Highest Score : ", highest_score)
```

```python
average_score = df['Score'].mean() ## -- .mean(): Seçilen sütundaki tüm sayısal değerlerin aritmetik ortalamasını hesaplar.
print("Average Score : ", average_score)
```

```python
print(df.describe())
## -- Count: Her sütundaki geçerli (non-null) değerlerin sayısı.
## -- Mean: Her sütunun ortalaması.
## -- Standard Deviation (std): Her sütundaki değerlerin standart sapması, veri setinin ne kadar yayıldığını gösterir.
## -- Min: Her sütundaki en küçük değer.
## -- 25%: Birinci çeyrek, verilerin %25'inin altında olduğu değer.
## -- 50% (Median): İkinci çeyrek, verilerin ortası/ortalaması.
## -- 75%: Üçüncü çeyrek, verilerin %75'inin altında olduğu değer.
## -- Max: Her sütundaki en büyük değer.
```

```python
import matplotlib.pyplot as plt ## -- MatPlotLib kütüphanesi & alias'ı (plt)
import numpy as np
```

```python
x = np.linspace(0,10,100) ## -- 0 ile 10 arasında eşit aralıklarla 100 adet değer üretir. Bu değerler X ekseninin noktalarını temsil eder.
y = np.sin(x) ## -- X değerlerinin sinüsünü hesaplar ve bu değerler y eksenini temsil eder.

plt.plot(x, y) ## -- X ve Y değerlerini kullanarak bir çizgi grafiği oluşturur.
plt.title('Sine Graph') ## -- Sinüs Grafiği
plt.xlabel('X Axis') ## -- X Ekseni
plt.ylabel('Y Axis') ## -- Y Ekseni
plt.grid(True) ## -- Grafiğe ızgara ( siyah yatay-dikey çizgiler ) ekler, bu da değerleri daha kolay okumayı sağlar.
plt.show() ## -- Grafiği görüntüler.
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.rand(50) ## -- 0 ile 1 arasında rastgele 50 değer üretir.
y = np.random.rand(50) ## -- Bu değerler x ve y eksenlerindeki nokta konumlarını belirler.
colors = np.random.rand(50) ## -- Her nokta için random bir renk belirlemek amacıyla 50 rastgele değer üretir.
sizes = 1000 * np.random.rand(50) ## -- Noktaların büyüklüğünü temsil eder.

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='inferno') ## cmap = PLASMA, INFERNO, MAGMA, CIVIDIS , VIRIDIS
## -- plt.scatter(): x ve y değerlerini kullanarak  dağılım grafiği çizer.
## -- c=colors: Her noktanın rengini colors dizisindeki değerlere göre ayarlar.
## -- s=sizes: Her noktanın boyutunu sizes dizisindeki değerlere göre ayarlar.
## -- alpha=0.5: 0.5 değeri, noktaların yarı saydam olmasını sağlar.
## -- cmap='inferno': Renk haritasını belirtir.
plt.title('Scatter Plot')
plt.xlabel('X Axius')
plt.ylabel('Y Axius')
plt.colorbar()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D ## --  3D grafikler oluşturmak için gerekli olan Axes3D modülünü çağırır.
```

```python
fig = plt.figure() ## -- Yeni bir figür  oluşturur.
ax = fig.add_subplot(111, projection='3d') ## --  Figürde 3D subplot oluşturur. 111 ifadesi, 1 satır, 1 sütunda tek bir grafik olduğunu belirtir.

x = np.linspace(-5, 5, 100) ## -- x ekseninde -5 ile 5 arasında 100 eşit aralıklı değer üretir.
y = np.linspace(-5, 10, 50) ## -- y ekseninde -5 ile 5 arasında 50 eşit aralıklı değer üretir.
X, Y = np.meshgrid(x, y) ## -- x ve y değerlerini mesh oluşturacak şekilde ayarlar.Yüzeyin her noktasını tanımlamak için gerekli bir yapıdır.
Z = np.sin(np.sqrt(X**2 + Y**2)) ## -- X ve Y değerlerini kullanarak her bir noktada sinüs fonksiyonunu hesaplar.

ax.plot_surface(X, Y, Z, cmap='plasma')

ax.set_title('3D Surface Plot Example')  ## -- Grafiğğe başlık ekler.
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
## -- farklı renk paletlerinin görünümüdür. ya da >>>> https://matplotlib.org/stable/users/explain/colors/colormaps.html
def plot_colormaps():
    colormaps = plt.colormaps()
    n = len(colormaps)
    fig, axs = plt.subplots(n//2, 2, figsize=(15, 2*(n//2)))
    for ax, cmap_name in zip(axs.ravel(), colormaps):
        cmap = plt.get_cmap(cmap_name)
        img = np.linspace(0, 1, 256).reshape(1, -1)
        ax.imshow(img, aspect='auto', cmap=cmap)
        ax.set_title(cmap_name, fontsize=8)
        ax.set_axis_off()
    plt.show()

plot_colormaps()
```
